```python
from sklearn.datasets import load_wine
```

```python
?load_wine
```

```python
data = load_wine(as_frame=True)
```

```python
data["frame"].head()
```

```python
data["frame"].info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data["data"]
```

```python
y = data["target"]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
?LogisticRegression
```

```python
clf = LogisticRegression(random_state=1, max_iter=10000, C=0.5, solver='sag')
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
?accuracy_score
```

```python
accuracy_score(y_true=y_test, y_pred=y_pred)
```

```python
clf.n_iter_
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
?DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=1)
```

```python
params_dict={
    "criterion": ["entropy", "gini"],
    "splitter": ["random", "best"],
    "min_samples_leaf": [2, 4, 6, 8, 10],
    "max_depth": [3,4,5,6]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
?GridSearchCV
```

```python
gs = GridSearchCV(estimator=clf, param_grid=params_dict, cv=4)
```

```python
gs.fit(X_train, y_train)
```

```python
y_pred1 = gs.predict(X_test)
```

```python
accuracy_score(y_true=y_test, y_pred=y_pred1)
```

```python
gs.best_estimator_
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ad = AdaBoostClassifier(random_state=1)
```

```python
params_dict={
    "n_estimators": [100, 500, 1000],
    "learning_rate": [0.5, 1, 2]
}
```

```python
gad = GridSearchCV(estimator=ad, param_grid=params_dict, cv=4)
```

```python
gad.fit(X_train, y_train)
```

```python
y_pred2 = gad.predict(X_test)
```

```python
accuracy_score(y_true=y_test, y_pred=y_pred2)
```

```python
gad.best_estimator_
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
?VotingClassifier
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier
```

```python
vt = VotingClassifier(
    estimators=[
        ("bc", BaggingClassifier(random_state=1)),
        ("rf", RandomForestClassifier(random_state=1)),
        ("gb", GradientBoostingClassifier(random_state=1))
    ]
)
```

```python
vt.fit(X_train, y_train)
```

```python
y_pred3 = vt.predict(X_test)
```

```python
accuracy_score(y_true=y_test, y_pred=y_pred3)
```
